### AnÃ¡lise Exploratoria dos dados

```python

data = []

for i, vec in enumerate(os.listdir('outputs/features/features_by_model')):
    pred_path = os.path.join('outputs/features/features_by_model', vec)
    vector = np.load(pred_path)
    
    # Add a string prefix to each element
    prefix = f"{i}."
    vector_str = [prefix + str(v) for v in vector]
    
    data.append(vector_str)
```

```python
vectors_np = np.hstack(data)
```

```python
df = pd.read_parquet('data/tce.parquet')
```

```python
df['cluster'] = vectors_np
```

### Valor Empenhado entre os clusters

```python
condition = df['Vlr_Empenhado'] < 0 # retirar do dataset os Vlr_Empenhados negativos, pois eles representam correÃ§Ãµes de empenhos antigos
data = df.drop(df[condition].index)

valor_empenhado = data[['Vlr_Empenhado','cluster']]

grupo_mean = valor_empenhado.groupby('cluster')['Vlr_Empenhado'].mean()
grupo_std = valor_empenhado.groupby('cluster')['Vlr_Empenhado'].std().fillna(0)
```

```python
import seaborn as sns

clusters_to_plot = ['40.0', '40.1', '40.2', '40.3', '40.4', '40.5']
plt.figure(figsize=(6,4))
sns.barplot(x=clusters_to_plot, y=grupo_mean.loc[clusters_to_plot].values)
plt.tight_layout()
plt.show()
```
